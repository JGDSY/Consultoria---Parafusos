```python
import pandas as pd

# Ler o arquivo CSV com a primeira linha como cabeçalho
dados = pd.read_csv('arquivo_modificado.csv')

# Limpar os nomes das colunas
dados.columns = dados.columns.str.strip()
dados.columns = dados.columns.str.replace(' ', '_')

# Visualizar os primeiros registros do dataframe
print(dados.head())
```

```python
dados = dados.astype({"phantom": "string"})
```

```python
# Calcular a média da coluna 'Qtde' agrupada por 'phantom' e 'PN'
media_por_phantom_e_pn = dados.groupby(['phantom', 'PN'])['Qtde'].mean()

# Exibir a média por phantom e PN
print(media_por_phantom_e_pn)
```

```python
# Calcular a média e a variância da coluna 'Qtde' por cada valor distinto da coluna 'vin'
media_var_por_vin = dados.groupby('vin')['Qtde'].agg(Media=lambda x: x.mean(), Variancia=lambda x: x.var()).reset_index()

# Visualizar o resultado da média e variância por vin
print(media_var_por_vin)

# Calcular a média e a variância da coluna 'Qtde' por cada tipo de 'PN'
media_var_por_PN = dados.groupby('PN')['Qtde'].agg(Media=lambda x: x.mean(), Variancia=lambda x: x.var()).reset_index()

# Visualizar o resultado da média e variância por PN
print(media_var_por_PN)
```

```python
# Calculando outras estatísticas descritivas

# Mediana
mediana_por_vin = dados.groupby('vin')['Qtde'].median().reset_index()
mediana_por_PN = dados.groupby('PN')['Qtde'].median().reset_index()
# Imprimindo as estatísticas descritivas adicionais
print("\nOutras estatísticas descritivas:")
print("\nMediana por 'vin':")
print(mediana_por_vin)
print("\nMediana por 'PN':")
print(mediana_por_PN)
# Percentis (25º, 50º e 75º)
percentis_por_vin = dados.groupby('vin')['Qtde'].quantile([0.25, 0.5, 0.75]).unstack().reset_index()
percentis_por_PN = dados.groupby('PN')['Qtde'].quantile([0.25, 0.5, 0.75]).unstack().reset_index()
print("\nPercentis por 'vin':")
print(percentis_por_vin)
print("\nPercentis por 'PN':")
print(percentis_por_PN)

# Desvio padrão
desvio_padrao_por_vin = dados.groupby('vin')['Qtde'].std().reset_index()
desvio_padrao_por_PN = dados.groupby('PN')['Qtde'].std().reset_index()
print("\nDesvio padrão por 'vin':")
print(desvio_padrao_por_vin)
print("\nDesvio padrão por 'PN':")
print(desvio_padrao_por_PN)
```

```python
# Coeficiente de variação
coef_var_por_vin = (dados.groupby('vin')['Qtde'].std() / dados.groupby('vin')['Qtde'].mean()).reset_index()
coef_var_por_PN = (dados.groupby('PN')['Qtde'].std() / dados.groupby('PN')['Qtde'].mean()).reset_index()
print("\nCoeficiente de variação por 'vin':")
print(coef_var_por_vin)
print("\nCoeficiente de variação por 'PN':")
print(coef_var_por_PN)
# Intervalo interquartil (IQR)
IQR_por_vin = dados.groupby('vin')['Qtde'].quantile(0.75) - dados.groupby('vin')['Qtde'].quantile(0.25)
IQR_por_PN = dados.groupby('PN')['Qtde'].quantile(0.75) - dados.groupby('PN')['Qtde'].quantile(0.25)
print("\nIntervalo interquartil (IQR) por 'vin':")
print(IQR_por_vin)
print("\nIntervalo interquartil (IQR) por 'PN':")
print(IQR_por_PN)
# Skewness
skewness_por_vin = dados.groupby('vin')['Qtde'].skew().reset_index()
skewness_por_PN = dados.groupby('PN')['Qtde'].skew().reset_index()

# Kurtosis
kurtosis_por_vin = dados.groupby('vin')['Qtde'].apply(pd.Series.kurtosis).reset_index()
kurtosis_por_PN = dados.groupby('PN')['Qtde'].apply(pd.Series.kurtosis).reset_index()
```
